```python
import seaborn as sns
import pandas as pd
import numpy as np
ship= sns.load_dataset('titanic')
ship.head()
```

```python
sns.boxplot(x='sex',y='fare', data=ship)
```

```python
sns.boxplot(x='class',y='fare', data=ship)
```

```python
sns.boxplot(x='survived',y='age',showmeans=True, data=ship)
```

```python
sns.boxplot(x='survived',y='age',showmeans=True,
            meanprops={'marker':'+',
                     "markersize":"12",
                     'markeredgecolor':'red'},
            data=ship)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ship= sns.load_dataset('titanic')
sns.boxplot(x='survived',y='age',showmeans=True,
            meanprops={'marker':'+',
                     "markersize":"12",
                     'markeredgecolor':'red'},
            data=ship)
plt.xlabel('How many survived',size=14, weight='bold')
plt.ylabel('Age',size=14, weight='bold')
plt.title('box plot of ship', size=14, weight='bold')
plt.show()
```
